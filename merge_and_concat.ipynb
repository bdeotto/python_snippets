```python
import pandas as pd
```

```python
def restart():
    """ Datasets para ilustración de merge y concat. 
    """
    
    global df1,df2
    
    df1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6]},index=[500,501,502])
    df2=pd.DataFrame({'A':[1,1,2,4],'B':[4,4,None,7]},index=[500,501,502,503])
    display(df1, df2)
    
    return 
```

```python
# Concat: # default join='outer' # default: cols+indice # Genera duplicados   # Conserva índices
# Merge:  # default how='inner'  # default: cols        # No genera duplicados # Default: Resetea índice
restart()
pd.concat([df1,df2])
```

```python
# Concat: # default join='outer' # default: cols+indice # Genera duplicados    # Conserva índices
# Merge:  # default how='inner'  # default: cols        # No genera duplicados # Default: Resetea índice
restart()
concat=pd.concat([df1,df2])
display(concat,df1.merge(df2,how='outer'))
```

```python
# Merge conservando índices originales: 
restart()
for df in [df1,df2]: 
    df.reset_index(inplace=True)
merge=df1.merge(df2,how='outer')
merge.set_index('index',inplace=True)
merge.sort_index(inplace=True)
merge
```

```python
restart()
concat=pd.concat([df1,df2]).drop_duplicates()
concat # drop_duplicates elimina duplicados con distinto índice
```

```python
# añadir a df1 la segunda col de df2 cuando df1['B']=df2['A']
restart()
df2.rename(columns={'B':'C','A':'B'},inplace=True)
df2
```
